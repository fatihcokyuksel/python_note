## PYTHON CLASSES and OBJECTS

**What is a Class?**
A class is a blueprint for creating objects.
It defines:

Attributes → what the object has

Methods → what the object can do

Creating a class does not create an object.

```python
#this class exists, but no object is created yet.
class Person:
    pass
```

**Creating Objects(Instances)**  
An object (or instance) is created by calling the class like a function.  
Each object:
- Has its own memory
- Can store different data

```python
p1 = Person()
p2 = Person()
```

**__init__ Method (Constructor)**  
The ***__init__*** method runs automatically when an object is created. It is used to initialize attributes.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
p = Person("Ali", 20)
print(p.name)   
print(p.age)    
```

**self** keyword
- Refers to the current object
- Used to access attributes and methods

**Instance Attributes vs Class Attributes**

Instance Attributes:  
- Belong to one object
- Defined using self

Class Attributes:  
- Belong to the class
- Shared by all objects

```python
class Dog:
    species = "Canine"      # class attribute

    def __init__(self, name):
        self.name = name    # instance attribute
```

```python
d1 = Dog("Max")
d2 = Dog("Rocky")

print(d1.species) 
print(d2.species)  
print(d1.name)
print(d2.name)
```

**Methods**  
Methods are functions inside a class.
- They always take **self** as the first parameter.

```python
class Counter:
    def __init__(self):
        self.value = 0

    def increase(self):
        self.value += 1

    def decrease(self):
        self.value -= 1
```

```python
c = Counter()
c.increase()
print(c.value) 
c.decrease()
print(c.value)
```

**Inheritance**
Inheritance allows a class to reuse another class. It models an "is-a" realationship.

```python
class Animal:
    def speak(self):
        return "Sound"
```

```python
class Cat(Animal):
    def speak(self):
        return "Meow"

class Dog(Animal):
    def speak(self):
        return "Hav Hav"
```

```python
cat = Cat()
dog = Dog()
print(cat.speak())
print(dog.speak())
```

**super()**  
Used to call the parent class constructor or methods.

```python
class Animal:
    def __init__(self, name):
        self.name = name
```

```python
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) #parent constructor
        self.breed = breed
```

```python
d = Dog("Max", "Golden")
print(d.name)
print(d.breed)
```

```python
#you can use super() outside of the constructor too
class Vehicle:
    def start(self):
        print("Vehicle started")
```

```python
class Car(Vehicle):
    def start(self):
        super().start()
        print("Car is ready to drive")
```

```python
c = Car()
c.start()
```

**Polymorphism**  
Different objects respond to the samae method name differently.

```python
class Dog:
    def sound(self):
        return "Bark"

class Bird:
    def sound(self):
        return "Chirp"

def make_sound(obj):
    print(obj.sound())
```

```python
make_sound(Dog())
make_sound(Bird())
```

**Encapsulation**  
It protects internal data from direct access. "__"

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
```

```python
acc = BankAccount(500)
print(acc.__balance)
```
