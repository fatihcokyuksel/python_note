# RECURSION #

Recursion is when a function calls itself to solve a smaller version of the same problem. Every recursive function must have a base case (stop condition) and a recursive case (call with smaller input).


```python
def countdown(n):
    if n == 0:
        print("done!")
        return
    print(n)
    countdown(n-1)

countdown(5)
```

```python
def factorial(n):
    if n == 0: #base case
        return 1
    return n * factorial(n - 1) #recursive case

print(factorial(5))
```

```python
#recursion with lists
def sum_list(lst):
    if len(lst) == 0:
        return 0
    return lst[0] + sum_list(lst[1:])
def recursive_max(lst):
    if len(lst) == 1:
        return lst[0]
    return max(lst[0], recursive_max(lst[1:]))

print(sum_list([1, 2, 3, 4, 5]))
print(recursive_max([1, 2, 3, 4, 5]))
```
